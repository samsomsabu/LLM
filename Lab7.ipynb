# Load the Dataset

```python
import pandas as pd

# Load your healthcare dataset
data = pd.read_csv('/content/train.csv')

data.head()
```

```python
texts = data['Track Name']
```

# Tokenize the Text

```python
import nltk
nltk.download('punkt')

# Tokenize the texts using NLTK
tokenized_texts = [nltk.word_tokenize(text) for text in texts]
```

# Create a Vocabulary

```python
from collections import defaultdict

# Create a vocabulary dictionary
vocab = defaultdict(int)
current_id = 1

for tokens in tokenized_texts:
    for token in tokens:
        if token not in vocab:
            vocab[token] = current_id
            current_id += 1

# Convert tokenized texts into sequences of IDs
encoded_texts = [[vocab[token] for token in tokens] for tokens in tokenized_texts]
```

# Padding and Truncation

```python
max_length = 50  # Set a maximum sequence length

# Pad or truncate sequences
padded_texts = [tokens[:max_length] + [0] * (max_length - len(tokens)) if len(tokens) < max_length else tokens[:max_length] for tokens in encoded_texts]
```
